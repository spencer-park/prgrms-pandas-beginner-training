## 비트연산자(Bitwise operators)를 이용한 다중 조건문(Multiple Conditions)
- `mask1 & mask2`를 사용합시다.
- `&` 는 and 연산을 한다.
- `|` 는 or, `~`는 not
- https://pandas.pydata.org/docs/user_guide/indexing.html#boolean-indexing
- https://pandas.pydata.org/docs/whatsnew/v0.19.0.html#logical-operators
    - 버전에 따라 동작이 바뀔 수 있음을 안내하고 있다.

```python
# 예제 데이터
x = pd.Series([True, False, True, False])
y = pd.Series([True, True, False, False])
```

```python
# and 연산
x & y
```


- 출처 : https://introcs.cs.princeton.edu/java/71boolean/

```python
# or 연산
x | y
```

```python
# not 연산
~x
```
