## 데이터 가공
1. Date열를 인덱스로 설정
2. Date열을 날짜 데이터(시계열 데이터)로 변환
    - Index >>> DatetimeIndex
    - 문자열(Object) >>> 특정 시간(Timestamp) >>> 기간(Period)

```python
# 데이터 로드
langs = pd.read_csv('Popularity of Programming Languages from 2004 to 2023.csv',
                   index_col='Date',
                   parse_dates=['Date'])
langs
```

```python
langs.index[0]
```

```python
# Timestamp >>> Period
langs.index = langs.index.to_period('m')
```

```python
langs
```

```python
langs.index[0]
```

## bar chart race 활용

```python
# 적은 값으로 일단 테스트
data = langs.iloc[:10]
data
```

```python
bcr.bar_chart_race(data)
```

## bar chart race 세부설정

```python
bcr.bar_chart_race(data, n_bars=10, figsize=(6, 4), sort='desc', period_length=600, title='Popular Programming Langs')
```

- https://www.dexplo.org/bar_chart_race/tutorial/#plot-properties
- n_bars : 기본적으로 모든 열이 그려짐. 개수 제한하려면 n_bars에 자연수 입력
- figsize: 시각화된 그래프의 크기를 설정하는 인자. 기본값 (6, 3.5).
- dpi: 해상도. 인치당 점의 수를 제어하는 인자.  dots per square inch(DPI)로 그래프의 해상도를 조. 기본값 144
- label_bars: 막대 그래프의 값을 텍스트로 표시할지 여부를 설정. 기본값 True
- title: 그래프의 제목을 설정. 문자열.
- sort : 정렬. asc, desc
- preiod_length : 주기당 시간을 변경. 기본적으로 500밀리초(0.5초)로 설정.
- steps_per_period : 프레임. 기본값 10. 클수록 부드러움.
